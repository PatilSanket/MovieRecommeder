```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
```

```python
credits=pd.read_csv(r'./Moviedata/tmdb_5000_credits.csv')
movies=pd.read_csv(r'./MovieData/tmdb_5000_movies.csv')
```

```python
credits.head(10)
```

```python
movies.head(10)
```

```python
credits.rename(columns={'movie_id':'id'},inplace=True)
credits.head()
```

```python
Total_movies = movies.merge(credits,on='id')
Total_movies.head(2)
```

```python
tfidf=TfidfVectorizer(stop_words='english',analyzer='word')
```

```python
Total_movies['overview']=Total_movies['overview'].fillna('')
tfidf_matrix=tfidf.fit_transform(Total_movies['overview'])
```

```python
print("TFIDF matrix shape is :",format(tfidf_matrix.shape))
print("Movies dataframe shape is :", format(Total_movies.shape))
```

```python
i=pd.Series(Total_movies.index,index=Total_movies['original_title'])
```

```python
def get_recommendations(title,similarity = linear_kernel(tfidf_matrix,tfidf_matrix)):
    idx=i[title]
    similar=list(enumerate(similarity[idx]))
    similar=sorted(similar,key=lambda x:x[1],reverse=True)
    similar=similar[1:6]
    movie_idx=[i[0] for i in similar]
    return movie_idx
```

```python
movie_index=get_recommendations('Toy Story')
for movie in movie_index:
    print(Total_movies['original_title'].iloc[movie])
```
